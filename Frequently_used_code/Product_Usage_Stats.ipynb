```python
import pandas as pd
```

```python
sandbox_products = '~/sandbox-products.csv'
```

```python
sandbox_df = pd.read_csv(sandbox_products).fillna('')
```

```python
usage_stats = '~/Files_20200401_20200430.csv'
```

```python
usage_df = pd.read_csv(usage_stats)
```

```python
s3_sandbox_df = pd.DataFrame(sandbox_df[sandbox_df['location'].str.startswith('s3://dea-public-data/')]).reset_index()
```

```python
s3_prefix = s3_sandbox_df['location'].str.replace('s3://dea-public-data/','')
```

```python
prefix_usage = []
for prefix in s3_prefix:
    product_usage = usage_df[usage_df['value'].str.contains(prefix+'/')]['hits'].sum()
    prefix_usage.append((prefix,product_usage))
```

```python
usage_cum_df = pd.DataFrame(prefix_usage, columns = ['prefix','total_usage'])
```

```python
usage_by_product = pd.concat([s3_sandbox_df,usage_cum_df],axis=1).sort_values('total_usage',ascending=False)[['product','location','total_usage']]
```

```python
usage_by_product.to_csv('product_usage_dea_public_data.csv',index=False)
```
