# Combining woody and not woody training sets from Gluevis into a ballanced training set for machine learning 

This notebook takes two CSVs from Gluevis, one with the spectral caracteristics for Trees and the over with the spectral characteristics for Not-Trees (should represent a combination of urban, grassland, cropland and water). the ratio between trees/not-trees is calculated and used to build a dataset where both classes are equaly represented. 
This combined set can then randomly sub-sampled to produce a smaller subset  or just shuffled to randomise the order of samples. 

```python
import pandas as pd
import numpy as np
```

```python
training_file_trees = '/g/data/u46/users/ck9738/Datasets/ML_for_smad_green/Glueviz_datasets/aus_alps_traning_11_trees.csv'

training_file_nottrees = '/g/data/u46/users/ck9738/Datasets/ML_for_smad_green/Glueviz_datasets/aus_alps_traning_11_not.csv'
save_path = '/g/data/u46/users/ck9738/Datasets/ML_for_smad_green/Glueviz_datasets/aus_alps_traning_11_small_even.csv'
```

```python
#Open file and check data exists
trees = pd.read_csv(training_file_trees)
not_trees = pd.read_csv(training_file_nottrees)
trees
```

```python
not_trees
#Open file and check data exists
```

```python
#add a colum for the classification called 'class' and vive it the value of 1 for Trees and 0 for Not Trees
trees['class'] = pd.Series(1, index=trees.index)
not_trees['class'] = pd.Series(0, index=not_trees.index)

#calulate the ration of trees to not trees presant
tree_num = len(trees.index)
not_num = len(not_trees.index)
ratio = tree_num/not_num
ratio
```

```python
#use calculated ratio to sub-sample an equal amount of each class

smaller_not = not_trees.sample(frac=ratio)
print(len(smaller_not.index),tree_num)
```

```python
#combine into one ballanced dataset

big_panda = trees.append(smaller_not, ignore_index=True)
big_panda
```

```python
smaller_subset = big_panda.sample(frac=0.4) #Change Frac toget a smaller sub-sample if you have too many datapoints. 
                                            #2 million is too many
smaller_subset
```

```python
#check that all values are finite 
np.all(np.isfinite(smaller_subset))
```

```python
#save to file 

smaller_subset.to_csv(save_path)
```
