```python

#load requiered modules 
import numpy as np
import rasterio
#from matplotlib import pyplot as plt #don't need unless want to plot
import xarray as xr
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt

from scipy import ndimage
from scipy.ndimage import measurements
from scipy import stats

import os

#set up filepath due to folder re-arrangement 
os.chdir('/g/data/r78/ck9738/')
b=0
```

```python
matrix_dataset = xr.Dataset()
```

```python
for i in os.listdir('/g/data/r78/ck9738/Matrixes'):
    
    lable = str(i)
    lable = lable[22:26]

    #read each file in above folder into a dataframe then into a numpy array?
    panda_table=pd.read_csv('/g/data/r78/ck9738/Matrixes/'+i,index_col=0)
    apple_pie=xr.DataArray(panda_table)
    matrix_dataset[lable]=apple_pie
    print(lable)
```
